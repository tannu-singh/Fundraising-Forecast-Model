```python
import csv
import pandas as pd
```

```python
dg= pd.read_csv('Attrition_Gift_Amount.csv')
dg.head()
```

```python
dg= dg[(dg[['2009','2010', '2011', '2012', '2013', '2014','2015']] != 0).all(axis=1)]
dg.shape
```

```python
dfr= pd.read_csv("Attrition_Frequency_Flag.csv")
dfr.head()
```

```python
dfr= dfr[(dfr[['2009','2010', '2011', '2012', '2013', '2014','2015']] != 0).all(axis=1)]
dfr.shape
```

```python
dbf= pd.read_csv("Attrition_Binary_Flag.csv")
dbf.head()
```

```python
dbf= dbf[(dbf[['2009','2010', '2011', '2012', '2013', '2014','2015']] != 0).all(axis=1)]
dbf.shape
```

```python
dc= pd.read_csv("OLV_DS_001_Decesased.csv")
dc.head()
```

```python
duplicates_dfr = set(dfr.index).intersection(dc.index)
dfr = dfr.drop(duplicates_dfr, axis=0)
dfr.shape
```

```python
duplicates_dg = set(dg.index).intersection(dc.index)
dg = dg.drop(duplicates_dg, axis=0)
dg.shape
```

```python
dg.head()
```

```python
duplicates_db = set(dbf.index).intersection(dc.index)
dbf = dbf.drop(duplicates_db, axis=0)
dbf.shape
```

```python
dfs= [dfr, dg, dbf]
```

```python
from functools import reduce
df_final = reduce(lambda left,right: pd.merge(left,right,on='Constituent ID'), dfs)
```

```python
df_final.head()
```

```python
df_final.shape
```

```python
df_final.columns
```

```python
ddf=  dfr.set_index('Constituent ID').stack().reset_index()
ddf = ddf.rename(columns={'level_1':'Year', 0:'Frequency'})
ddf.head()
```

```python
ddb=  dbf.set_index('Constituent ID').stack().reset_index()
ddb = ddb.rename(columns={'level_1':'Year', 0:'Attrition'})
ddb.head()
```

```python
ddon=  dg.set_index('Constituent ID').stack().reset_index()
ddon = ddon.rename(columns={'level_1':'Year', 0:'Donation Amount'})
ddon.head()
```

```python
df_features= [ddf, ddb, ddon]
from functools import reduce
df_merge = reduce(lambda left,right: pd.merge(left,right,on='Constituent ID'), df_features)
df_merge.head()
```

```python
df_merge.columns
```

```python
del df_merge['Year_y']
del df_merge['Year_x']
df_merge.to_csv('Filtered_Constituents.csv')
df_merge.head()
```
