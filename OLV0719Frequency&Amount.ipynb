```python
import pandas as pd

df= pd.read_csv('OLV0719FrequencyAttrition.csv')
da= pd.read_csv('OLV0719AmountAttrition.csv')
dc= pd.read_csv("OLV_DS_001_Decesased.csv")
df.head(), da.head(), dc.head()
```

```python
duplicates_da = set(da.index).intersection(dc.index)
da = da.drop(duplicates_da, axis=0)
da.shape
da.head()
```

```python

duplicates_df = set(df.index).intersection(dc.index)
df = df.drop(duplicates_df, axis=0)
df.head()
```

```python
df.shape
```

```python
df= df[(df[['2009','2010', '2011' , '2012', '2013', '2014', '2015']] != 0).all(axis=1)]
#da= da[(da[['2009','2010', '2011', '2012', '2013', '2014','2015']] != 0).all(axis=1)]
```

```python
df.head()
```

```python
dfb=  df.set_index('Constituent ID').stack().reset_index()
dfb = dfb.rename(columns={'level_1':'Year', 0:'Frequency'})
dfa=  da.set_index('Constituent ID').stack().reset_index()
dfa = dfa.rename(columns={'level_1':'Year', 0:'Total Amount'})
```

```python
dfa.head()
```

```python
df_features= [dfb, dfa]
from functools import reduce
df_merge = reduce(lambda left,right: pd.merge(left,right,on='Constituent ID'), df_features)


df_merge.head()
```

```python
df_merge['Total Donation']= df_merge.groupby(["Constituent ID"])["Total Amount"].transform(sum)
df_merge['Total Frequency']= df_merge.groupby(["Constituent ID"])["Frequency"].transform(sum)
df_merge.head()
```

```python
del df_merge['Frequency']
del df_merge['Year_y']

```

```python
del df_merge['Total Amount']
```

```python
d= df_merge.drop_duplicates()
d.to_csv('OLV0719Frequency&AmountFeature.csv')
```
